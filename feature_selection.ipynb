## F1 Score - Thyroid

### Manual

```python
X = normalized_data_tyhroid.values
y = df1['Recurred'].values
```

```python
import numpy as np

def fisher_discriminant_ratio_thyroid(X, y):
    classes, counts = np.unique(y, return_counts=True)
    c = len(classes)  # Number of classes

    means = []
    variances = []
    proportions = []
    output = []  # Output for each column

    for i in range(X.shape[1]):
        means = []
        variances = []
        proportions = []

        for cls in classes:
            X_cls = X[y == cls, i]
            cls_mean = np.mean(X_cls)
            cls_variance = np.mean(np.var(X_cls))
            cls_proportion = len(X_cls) / len(X)

            means.append(cls_mean)
            variances.append(cls_variance)
            proportions.append(cls_proportion)

        # Calculate numerator and denominator
        numerator = 0
        for i in range(c):
            for j in range(i + 1, c):
                numerator += proportions[i] * proportions[j] * np.sum((means[i] - means[j]) ** 2)

        denominator = 0
        for i in range(c):
            denominator = np.sum(proportions[i] * variances[i])

        rfi = numerator / denominator
        output.append(rfi)

    return output

rfi = fisher_discriminant_ratio_thyroid(X, y)

print("Output:")
print(rfi)
```

```python
F1_tyhroid = 1 / (1 + max(rfi))
print(f"F1 score: {F1_tyhroid:.5f}")
```

```python
# Filter fitur ReliefF yang memiliki skor di atas threshold
relieff_selected_features = sorted_feature_scores_df_thyroid[sorted_feature_scores_df_thyroid['Score'] > F1_tyhroid]

# Filter fitur Lasso yang memiliki koefisien absolut di atas threshold
lasso_selected_features = feature_ranking_thyroid[feature_ranking_thyroid > F1_tyhroid]

# Filter fitur RandomForestClassifier yang memiliki tingkat penting di atas threshold
rf_selected_features = feature_importance_df_thyroid[feature_importance_df_thyroid['Importance'] > F1_tyhroid]

# Menampilkan fitur yang dipilih untuk masing-masing metode beserta nilainya
print("Fitur yang dipilih dengan ReliefF:")
print(relieff_selected_features)

print("\nFitur yang dipilih dengan Lasso:")
print(lasso_selected_features)

print("\nFitur yang dipilih dengan RandomForestClassifier:")
print(rf_selected_features)
```
